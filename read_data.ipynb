```python
import numpy as np
import pandas as pd
import random
```

```python
DATA_FILENAME = 'data/train_1.csv'
p = 0.7 # 1% of the data
```

```python
df = pd.read_csv(DATA_FILENAME,skiprows=lambda i: i>0 and random.random() > p)
df.shape
```

```python
df = df.fillna(0)
```

```python
df.head(5)
```

```python
def extract_info_page(row):
    page = row['Page']
    website_name = '.wikipedia.org_'
    i = page.find(website_name)
    name = page[:i-3] #skipping the 2 characters domain and the '_'
    domain = page[i-2:i] # characters domaine of size 2 before the .wikipedia.org
    acces_agent = page[i+len(website_name):].split('_') # the rest of the string is the access and agent splitted by a '_'
    access = acces_agent[0]
    agent = acces_agent[1]
    return name,domain,access,agent
```

```python
# so it put the new columns in the beginning
df.insert(0,'Name','')
df.insert(0,'Domain','')
df.insert(0,'Access','')
df.insert(0,'Agent','')
```

```python
df['Name'], df['Domain'],df['Access'],df['Agent'] = zip(*df.apply(extract_info_page,axis=1))
df.head(5)
```
